Feedback problem with simpleSBML

```python
import simplesbml
```

Must add species as modifiers if required

```python
model = simplesbml.SbmlModel()
model.addSpecies('$X0', 10)
model.addSpecies('S1', 0)
model.addSpecies('S2', 0)
model.addSpecies('S3', 0)
model.addSpecies('$X1', 0)
model.addParameter('VM1', 10)
model.addParameter('Keq1', 10)
model.addParameter('h', 10)
model.addParameter('V4', 2.5)
model.addParameter('KS4', 0.5)
spec = model.addSpecies('S4', 0)
re = model.addReaction(['X0'], ['S1'], '(VM1 * (X0 - S1/Keq1))/(1 + X0 + S1 +   S4^h)', rxn_id='J0')
re.addModifier(spec)
# spec = model.addSpecies('S8', 0)
# re = model.addReaction(['S1'], ['S2'], '(10 * S1 - 2 * S2) / (1 + S1 + S2 + S8)', rxn_id='J1')
# re.addModifier(spec)
model.addReaction(['S1'], ['S2'], '(10 * S1 - 2 * S2) / (1 + S1 + S2)', rxn_id='J1')
model.addReaction(['S2'], ['S3'], '(10 * S2 - 2 * S3) / (1 + S2 + S3)', rxn_id='J2')
model.addReaction(['S3'], ['S4'], '(10 * S3 - 2 * S4) / (1 + S3 + S4)', rxn_id='J3')
model.addReaction(['S4'], ['X1'], '(V4 * S4) / (KS4 + S4)', rxn_id='J4')
```

```python
import tellurium as te
r = te.antimonyConverter()
antimony_rep = r.sbmlToAntimony(model.toSBML())[1]
```

```python
temodel = te.loada(antimony_rep)
temodel.simulate(
    start=0,
    end=40,
    points=100
)
temodel.plot()
```
